```python
sequences = []

for rows in range(len(a)):
    for cols in range(len(a[rows])):
        words = a[rows][cols:cols+4]
        if not words:
            continue
        elif (len(words)!=4):
            continue
        sequences.append(words)
        


print("The Length of sequences are: ", len(sequences))
sequences = np.array(sequences)
print(sequences)
```
